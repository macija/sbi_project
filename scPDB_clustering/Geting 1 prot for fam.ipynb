```python
pdb_families = pd.read_csv("pdb_families.csv", header=None)
```

```python
unique_fams = pd.unique(pdb_families[ [1,2,3,4]].values.ravel('K'))
```

```python
pdb_families.iloc[815][0]
```

```python
fam_contents = {}
for fam in unique_fams:
    a = pdb_families.index[(pdb_families[0] == fam) |
                              (pdb_families[1] == fam) | 
                                (pdb_families[2] == fam) |
                                (pdb_families[3] == fam) |
                                (pdb_families[4] == fam)]
    fam_contents[fam] = [pdb_families.iloc[x][0] for x in a]
```

```python
fam_contents
```

```python
chosen_pdbs = []
for i in fam_contents:
    if len(fam_contents[i]) >0:
        chosen_pdbs.append(random.choice(fam_contents[i]))
    else:
        print(fam_contents[i])
        print(i)
```

```python
print(chosen_pdbs)
```

```python
with open("families_1rep.txt", 'w') as fout:
    for i in chosen_pdbs:
        fout.write(i+'\n')
fout.close()
```
